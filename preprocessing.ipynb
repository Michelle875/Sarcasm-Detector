# Preprocessing Data

```python
#libraries
import pandas as pd
import numpy as np
import re
```

```python
df_train = pd.read_csv('train.csv')
df_valid = pd.read_csv('valid.csv')
df_test = pd.read_csv('test.csv')
```

## remove uncessary whitespace 

```python
print(df_train.head())
df_train['text'] = df_train['text'].str.strip()
print(df_train.head())
```

## Tokenization


```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
df_train['text'] = df_train['text'].apply(word_tokenize)
df_valid['text'] = df_valid['text'].apply(word_tokenize)
df_test['text'] = df_test['text'].apply(word_tokenize)
print(df_train.head())
print(df_valid.head())
print(df_test.head())
print(df_train.iloc[85][0]) # this is an example that has elipsis 
```

## Lemmanization
Michelle - we should test with and without lemmanization to see what works best


```python
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()

# Convert POS tags for WordNet
def get_wordnet_pos(tag):
    if tag.startswith('J'): 
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN   # default

def lemmatize_tokens(tokens):
    pos_tags = nltk.pos_tag(tokens)
    return [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) 
            for word, tag in pos_tags]

df_train['text'] = df_train['text'].apply(lemmatize_tokens)
df_valid['text'] = df_valid['text'].apply(lemmatize_tokens)
df_test['text'] = df_test['text'].apply(lemmatize_tokens)

print(df_train.head())
```

## export data

```python
df_train.to_csv('train_preprocessed.csv', index=False)
df_valid.to_csv('valid_preprocessed.csv', index=False)
df_test.to_csv('test_preprocessed.csv', index=False)
```
